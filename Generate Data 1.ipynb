```python
import pandas as pd
import random
from faker import Faker
from random import randrange
from datetime import datetime
```

```python
# create a variety of data types

nr_of_customers = 1000

fake = Faker('de_DE')

customers = []

for customers_id in range(nr_of_customers):

    # Create transaction date 
    d1 = datetime.strptime(f'1/1/2021', '%m/%d/%Y')
    d2 = datetime.strptime(f'8/10/2021', '%m/%d/%Y')
    transaction_date = fake.date_between(d1, d2)

    #create customer's name
    name = fake.name()

    # Create gender
    gender = random.choice(["M", "F"])

    # Create email 
    email = fake.ascii_email()

    #Create city
    city = fake.city()

    #create product ID in 8-digit barcode
    product_ID = fake.ean(length=8)
    
    #create amount spent
    amount_spent = fake.pyfloat(right_digits=2, positive=True, min_value=1, max_value=100)

    customers.append([transaction_date, name, gender, email, city, product_ID, amount_spent])

customers_df = pd.DataFrame(customers, columns=['Transaction_date','Name', 'Gender','Email', 'City','Product_id', 'Amount_spent']) 
                
pd.pandas.set_option('display.max_columns', None)
print(customers_df)
```

```python
# Generate random data
# C has 10 random values between 1 and 100
# D has 10 random values following standard normal distribution having mean 0 and standard deviation

import pandas as pd
import numpy as np
np.random.seed(1)
data3 = pd.DataFrame({"C" : np.random.randint(low=1, high=100, size=10),
                     "D"  : np.random.normal(0.0, 1.0, size=10)
                     })
```

```python
data3
```

```python
# Check mean and standard deviation of normal distribution
np.round(np.std(np.random.normal(0.0, 1.0, size=1000)))
np.round(np.mean(np.random.normal(0.0,1.0, size=1000)))
```

```python
# Create catagorical variable
# X has 20 values from 1 to 4 (excluding 5) with replacement (i.e. repeated values).
# X1 has binary variables (0 / 1)

np.random.seed(1)
data4 =pd.DataFrame({"X"  : np.random.choice(range(1,5), 20, replace=True),
                     "X1" : np.where(np.random.normal(0.0, 1.0, size=20)<=0,0,1)})
```

```python
data4
```

```python
data4.X  = data4.X.astype("category")
data4.X1 = data4.X1.astype("category")
```
