```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Plot the training points
X, y = make_blobs(n_samples=300, n_features=2, cluster_std=.1
                  ,centers= [(1,5), (2,10)])
plt.scatter(X[:,0],X[:,1], c=y);
```
